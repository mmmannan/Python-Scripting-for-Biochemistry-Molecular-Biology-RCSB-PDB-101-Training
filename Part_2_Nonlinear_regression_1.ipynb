```python
# Create the filehandle for the csv file that contains your data
data_file = os.path.join("CSV_folder", "AP_kin.csv")
print(data_file)
```

```python
# Create a second dataframe with the substrate concentration as the first series.
# Note the syntax for this pandas function - the D and F are capitalized.
MM_df = pd.DataFrame()
MM_df['pNP (mM)'] = AP_kin_df.columns      ## pNP = p-nitrophenol
## 2nd_df_name[col_name in 2nd_df] = values
## 2nd_df_name[col_name in 2nd_df] = 1st_df.columns
MM_df
```

```python
slopes = []
# Calculate the slopes from each column in the AP_kin_df dataframe
for column in AP_kin_df.columns:
    slope, _, _, _, _ = stats.linregress(AP_kin_df.index, AP_kin_df[column])
    slopes.append(slope)
    
slopes
```

```python
MM_df["slopes"] = slopes
MM_df
```

```python
# Calculate the initial velocity
MM_df['initial_velocities'] = MM_df["slopes"]/0.015
MM_df
```

```python
# make pNP (mM) the index for the dataframe.
MM_df.set_index('pNP (mM)', inplace = True)
MM_df
```

```python
#  write it to a csv file in CSV_folder directory.
# to write dataframe to a csv file use .to_csv()
output_file = os.path.join('CSV_folder', 'MM_df.csv')
MM_df.to_csv(output_file)
```
