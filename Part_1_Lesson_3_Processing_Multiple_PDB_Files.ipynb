```python
for pdb_file in file_names:
    
    # Get the pdb id
    f_name = os.path.basename(pdb_file)
    pdb_split = f_name.split(".")
    pdb_id = pdb_split[0]
    
    # Read the data
    with open(pdb_file, 'r') as outfile:
        data = outfile.readlines()
    
    for line in  data:
        if 'RESOLUTION.' in line:
            res_line = line
            words = res_line.split()
            resolution = float(words[3])
            print(pdb_id, resolution)
```

```python
for pdb_file in file_names:
    
    # Get the pdb id
    f_name = os.path.basename(pdb_file)
    pdb_split = f_name.split(".")
    pdb_id = pdb_split[0]
    
    # Read the data
    with open(pdb_file, 'r') as outfile:
        data = outfile.readlines()
    
    for line in  data:
        if 'RESOLUTION.' in line:
            res_line = line
            words = res_line.split()
            resolution = float(words[3])
            print(pdb_id, ":", resolution, "Angstroms")
```

```python
# Writing / Printing to a file
with open ("Resolutions.txt", "w+") as data_file:
    
    for pdb_file in file_names:
    
        # Get the pdb id
        f_name = os.path.basename(pdb_file)
        pdb_split = f_name.split(".")
        pdb_id = pdb_split[0]

        # Read the data
        with open(pdb_file, 'r') as outfile:
            data = outfile.readlines()

        for line in  data:
            if 'RESOLUTION.' in line:
                res_line = line
                words = res_line.split()
                resolution = float(words[3])
                data_file.write(F'{pdb_id} \t {resolution} \n')      # F string formatting
```

```python
with open ("Resolutions_2.txt", "w+") as data_file:
    
    for pdb_file in file_names:
    
        # Get the pdb id
        f_name = os.path.basename(pdb_file)
        pdb_split = f_name.split(".")
        pdb_id = pdb_split[0]

        # Read the data
        with open(pdb_file, 'r') as outfile:
            data = outfile.readlines()

        for line in  data:
            if 'RESOLUTION.' in line:
                res_line = line
                words = res_line.split()
                resolution = float(words[3])
                data_file.write(F'For the PDB ID {pdb_id}, the resolution is {resolution} in Angstroms.\n')
                # F string formatting
```
